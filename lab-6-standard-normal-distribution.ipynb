# Normal Distribution

```python
# normal distribution curve
mean = df['Current'].mean()
std = df['Current'].std(ddof=1)

# x-axis 
x_min = df['Current'].min()
x_max = df['Current'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color='#126782',
        linewidth=2)
plt.title('Normal Distribution of Current Values',
        font='Palatino Linotype',
        fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Probability Density')
plt.show()
```

```python
# histogram plot
plt.hist(df['Current'], bins=5,
        density=True,
        alpha = 0.8,
        color='#126782',
        edgecolor='#1C1F24')

plt.title('Normal Distribution of Current Values',
        font='Palatino Linotype',
        fontsize=12)
plt.xlabel('Current (mA)')
plt.ylabel('Frequency')

p = stats.norm.pdf(x, mean, std)
plt.plot(x, p,
        color='#FF6F61',
        linewidth=2)
plt.show()
```

# Standard Normal Distribution N(0,1)

```python
# standard normal distribution N(0,1)
# x-axis
x_max = df['Z-score'].max()
x_min = df['Z-score'].min()
x = np.linspace(x_min, x_max, 100) 

p = stats.norm.pdf(x, df['Z-score'].mean(), df['Z-score'].std(ddof=1))

plt.plot(x, p,
        color='#126782',
        linewidth=2)
plt.title('Standard Normal Distribution of Current Values',
        font='Palatino Linotype',
        fontsize=12)
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.show()
```
